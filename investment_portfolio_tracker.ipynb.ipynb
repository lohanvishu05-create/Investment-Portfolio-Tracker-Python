```python
import pandas as pd
import numpy as np

portfolio = {
    "Asset": ["Equity MF", "Debt MF", "Gold ETF", "Stocks"],
    "Invested_Amount": [200000, 100000, 50000, 150000],
    "Current_Value": [260000, 110000, 75000, 180000]
}

df = pd.DataFrame(portfolio)
df
```

```python
# Absolute return
df["Profit"] = df["Current_Value"] - df["Invested_Amount"]

# Return %
df["Return_%"] = (df["Profit"] / df["Invested_Amount"]) * 100

# Portfolio weight
total_value = df["Current_Value"].sum()
df["Weight_%"] = (df["Current_Value"] / total_value) * 100

df
```

```python
summary = {
    "Total Invested": df["Invested_Amount"].sum(),
    "Current Portfolio Value": df["Current_Value"].sum(),
    "Total Profit": df["Profit"].sum(),
    "Portfolio Return %": (df["Profit"].sum() / df["Invested_Amount"].sum()) * 100
}

summary
```

```python
import matplotlib.pyplot as plt

plt.pie(df["Weight_%"], labels=df["Asset"], autopct="%1.1f%%")
plt.title("Portfolio Allocation")
plt.show()
```
